```python
from collections import deque

# Graph definition
graph = {
    'A': ['B', 'C'],
    'B': ['D', 'E'],
    'C': ['F'],
    'D': [],
    'E': ['F'],
    'F': []
}
```

```python
# -------------------------
# Breadth First Search (BFS)
# -------------------------
def bfs(start):
    visited = []
    queue = deque([start])

    while queue:
        node = queue.popleft()
        if node not in visited:
            print(node, end=" ")
            visited.append(node)
            queue.extend(graph[node])

# -------------------------
# Run Traversals
# -------------------------
print("BFS Traversal:")
bfs('A')
```

```python
# -------------------------
# Depth First Search (DFS)
# -------------------------
visited = []

def dfs(node):
    if node not in visited:
        print(node, end=" ")
        visited.append(node)
        for neighbour in graph[node]:
            dfs(neighbour)

# -------------------------
# Run Traversals
# -------------------------

print("\nDFS Traversal:")
visited = []   # Reset visited before DFS
dfs('A')
```
