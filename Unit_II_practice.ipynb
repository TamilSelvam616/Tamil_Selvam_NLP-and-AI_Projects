```python
# =========================
# UNIT II – NLP LAB
# Lemmatization & POS Tagging
# =========================

import nltk
from nltk.tokenize import word_tokenize
from nltk import pos_tag
from nltk.stem import WordNetLemmatizer
from textblob import Word
import spacy

# Downloads (run once)
nltk.download('punkt')
nltk.download('wordnet')
nltk.download('averaged_perceptron_tagger')

# -------------------------
# 1. Lemmatization (Single Word)
# -------------------------
lemmatizer = WordNetLemmatizer()
word = "meeting"
print("Lemmatized Word:", lemmatizer.lemmatize(word, pos='v'))

# -------------------------
# 2. Lemmatize Multiple Words
# -------------------------
words = ["flew", "drawing", "ran", "drawn"]
for w in words:
    print(w, "→", lemmatizer.lemmatize(w, pos='v'))

# -------------------------
# 3. Lemmatization with POS
# -------------------------
sentence = "The dogs are running"
tokens = word_tokenize(sentence)
tags = pos_tag(tokens)

lemmatized = [
    lemmatizer.lemmatize(word, pos='v') if tag.startswith('V') else lemmatizer.lemmatize(word)
    for word, tag in tags
]

print(lemmatized)

# -------------------------
# 4. Task Sentence
# -------------------------
sentence = "I am studying books"
tokens = word_tokenize(sentence)
tags = pos_tag(tokens)

lemmatized = [
    lemmatizer.lemmatize(word, pos='v') if tag.startswith('V') else lemmatizer.lemmatize(word)
    for word, tag in tags
]

print(lemmatized)

# -------------------------
# 5. TextBlob Lemmatization
# -------------------------
word = Word("running")
print("TextBlob Lemma:", word.lemmatize("v"))

# -------------------------
# 6. POS Tagging (NLTK)
# -------------------------
text = "NLTK is a powerful library for natural language processing."
tokens = word_tokenize(text)
tags = pos_tag(tokens)

print("\nPOS Tagging Result:")
for word, tag in tags:
    print(word, ":", tag)

# -------------------------
# 7. POS Tagging Task Sentence
# -------------------------
sentence = "The quick brown fox jumps over the lazy dog."
tokens = word_tokenize(sentence)
tags = pos_tag(tokens)

for word, tag in tags:
    print(word, ":", tag)

# -------------------------
# 8. POS Tagging using spaCy
# -------------------------
nlp = spacy.load("en_core_web_sm")
text = "SpaCy is a popular natural language processing library."
doc = nlp(text)

print("\nspaCy POS Tags:")
for token in doc:
    print(token.text, ":", token.pos_)
```
